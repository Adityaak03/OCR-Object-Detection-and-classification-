```python
!pip install ultralytics
from ultralytics import YOLO
import cv2


plate_model = YOLO("/kaggle/input/li/other/default/1/numberplate714.pt")
smoke_model = YOLO("/kaggle/input/bestsmokemodel/other/default/1/smoke0.94")

cap = cv2.VideoCapture("/kaggle/input/hillvideo/hilldrive.mp4")
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)
out = cv2.VideoWriter("/kaggle/working/output_combined.mp4", cv2.VideoWriter_fourcc(*'mp4v'), fps, (width, height))

frame_count = 0

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    frame_count += 1
    print(f"Processing frame {frame_count}...")

    
    plate_results = plate_model(frame)[0]
    plate_boxes = plate_results.boxes.xyxy.cpu().numpy()

    for box in plate_boxes:
        x1, y1, x2, y2 = map(int, box)
        cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)  # Green
        cv2.putText(frame, "Plate", (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 
                    0.7, (0, 255, 0), 2)

  
    smoke_results = smoke_model(frame)[0]
    smoke_boxes = smoke_results.boxes.xyxy.cpu().numpy()

    for box in smoke_boxes:
        x1, y1, x2, y2 = map(int, box)
        cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 0, 255), 2)  # Red
        cv2.putText(frame, "Smoke", (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 
                    0.7, (0, 0, 255), 2)

   
    cv2.putText(frame, f"Frame: {frame_count}", (20, 30), cv2.FONT_HERSHEY_SIMPLEX, 
                0.8, (0, 255, 255), 2)

 
    out.write(frame)

cap.release()
out.release()
cv2.destroyAllWindows()
print("âœ… Video processing complete. Output saved as 'output_combined.mp4'")
```
